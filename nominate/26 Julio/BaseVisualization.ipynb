```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv('df_nominate.csv', index_col=0)
```

```python
df['names'] = df['names'].apply(lambda w: " ".join([l.capitalize() for l in " ".join(w.split(',')[::-1]).split(' ')]))
```

```python
color_range = ['#f94144', '#ffa600', '#8259b2', '#e41791', '#f9c74f', '#90be6d', '#277da1']
color_domain = ['Apruebo Dignidad', 'Nueva Constituci√≥n', 'Independientes', 'Lista del Pueblo','Lista del Apruebo','Pueblos Originarios','Vamos por Chile']
```

```python
alt.Chart(df).mark_point(filled=True, size=50, opacity=1).encode(x=alt.X('x', title='', scale=alt.Scale(domain=[-1.1, 1.1]), axis=None), 
                                  y=alt.Y('y', title='', scale=alt.Scale(domain=[-1.1, 1.1]), axis=None), 
                                  tooltip=[alt.Tooltip('names', title='Constituyente'), alt.Tooltip("party", title='Coalici√≥n')], 
                                  color=alt.Color('party', title='Coalici√≥n', scale=alt.Scale(domain=color_domain, range=color_range))).configure_axis(
    grid=False
)
```
